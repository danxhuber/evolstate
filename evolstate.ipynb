```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

from evolstate import *

# read in a bunch of stellar parameters
epic=ascii.read('k2c4_stparas.txt')

# get evolutionary states given Teff & R 
cl= evolstate(epic['Teff'],epic['Rad'],-99.+np.zeros(len(epic['Teff'])))

plt.ion()
plt.clf()
plt.semilogy(epic['Teff'],epic['Rad'],'.')
# main-sequence
plt.semilogy(epic['Teff'][cl == 0],epic['Rad'][cl == 0],'.')
# subgiants
plt.semilogy(epic['Teff'][cl == 1],epic['Rad'][cl == 1],'.')
# red giants. NB for this population some horizontal branch stars are cut-of
plt.semilogy(epic['Teff'][cl == 2],epic['Rad'][cl == 2],'.')
# NB this flags candidate cool main-sequence binaries
plt.semilogy(epic['Teff'][cl == 3],epic['Rad'][cl == 3],'.')
plt.xlim([8000,3000])
plt.ylim([0.3,100])
```

```python
# get evolutionary states given Teff & logg
cl= evolstate(epic['Teff'],np.zeros(len(epic['Teff'])),epic['logg'])
plt.ion()
plt.clf()
plt.semilogy(epic['Teff'],epic['logg'],'.')
plt.semilogy(epic['Teff'][cl == 0],epic['logg'][cl == 0],'.')
plt.semilogy(epic['Teff'][cl == 1],epic['logg'][cl == 1],'.')
plt.semilogy(epic['Teff'][cl == 2],epic['logg'][cl == 2],'.')
plt.xlim([8000,3000])
plt.ylim([5.5,1])
```
